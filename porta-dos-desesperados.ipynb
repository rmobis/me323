```python
import numpy as np
import matplotlib.pyplot as pyplot

size = 1000

# gera a lista de portas "corretas"
correct_doors = np.random.randint(1, 4, size = size, dtype = np.uint8)

# gera a lista de portas escolhidas inicialmente
first_choices = np.random.randint(1, 4, size = size, dtype = np.uint8)

# determina as escolhas corretas quando NO  feita a troca
correct_noswap = np.equal(correct_doors, first_choices)

# determina as escolhas corretas quando   feita a troca
correct_swap = np.invert(correct_noswap)

#
cumsum_noswap = np.cumsum(correct_noswap)
cumsum_swap = np.cumsum(correct_swap)

prop_cumsum_noswap = cumsum_noswap / (1 + np.indices(cumsum_noswap.shape)[0])
prop_cumsum_swap = cumsum_swap / (1 + np.indices(cumsum_noswap.shape)[0])

pyplot.figure(figsize=(16,9))
pyplot.title('Porta dos Desesperados')
pyplot.plot(prop_cumsum_noswap, label = 'No Trocar')
pyplot.plot(prop_cumsum_swap, label = 'Trocar')
pyplot.xlabel("# de Experimentos")
pyplot.ylabel("Probabilidade de Acerto")
pyplot.gca().set_ylim([0, 1])
pyplot.legend()
pyplot.show()
```
